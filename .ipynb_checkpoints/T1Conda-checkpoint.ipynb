# INSTALACION DEL ENTORNO (CONDA)
Conda es una potente herramienta de línea de comandos para la gestión de paquetes y entornos que se ejecuta en Windows, macOS y Linux, se puede utilizar con cualquier aplicación de terminal (o intérprete de comandos).

---

Para instalar conda, primero debes elegir el instalador adecuado. A continuación, se muestran los instaladores más populares disponibles actualmente:

1.  **Miniconda** Miniconda es un instalador mínimo proporcionado por Anaconda. Utilice este instalador si desea instalar la mayoría de los paquetes manualmente.

2.  **DISTRIBUCION ANACONDA** Anaconda Distribution es un instalador completo que incluye un conjunto de paquetes para la ciencia de datos, así como Anaconda Navigator, una aplicación con interfaz gráfica de usuario para trabajar con entornos conda.

   
3.  **MINI FORJA** Miniforge es un instalador mantenido por la comunidad conda-forge que viene preconfigurado para usarse con el canal conda-forge. Para obtener más información sobre conda-forge.



* **Linux:** [Instalador de Miniconda para Linux](https://docs.conda.io/en/latest/miniconda.html#linux-installers)
* **Windows:** [Instalador de Miniconda para Windows](https://docs.conda.io/en/latest/miniconda.html#windows-installers)
* **macOS:** [Instalador de Miniconda para macOS](https://docs.conda.io/en/latest/miniconda.html#macos-installers)

# Instalación de Conda (Miniconda)

Una vez descargado el archivo ejecuta (asegurate de que se la ultima version):

```python
bash ~/Miniconda3-latest-Linux-x86_64.sh
```

Una vez ejecutado el instalador en la terminal, sigue esta secuencia de interacción:

1.  **Revisión de Licencia** Presiona <kbd>Enter</kbd> para desplazarte y revisar los términos del acuerdo.
    
2.  **Aceptación del CLUF** Cuando el instalador se detenga, escribe `yes` y presiona <kbd>Enter</kbd> para aceptar el contrato de licencia.
    
3.  **Ubicación de los Archivos** El instalador sugerirá una ruta por defecto. Simplemente presiona <kbd>Enter</kbd> para confirmar.



##### Nuevamente en la termianl ingresa el siguiente comando para que la instalación surta efecto por completo:

```python
source ~/.bashrc
```

## Verificar la instalación

Verifica tu instalación abriendo la aplicación de terminal y ejecutando <code>conda list</code>. Si conda se ha instalado correctamente, aparecerá una lista de paquetes instalados:

```python
conda list
```

# CREACION DEL ENTORNO 

Los entornos en conda son espacios aislados y autónomos donde se pueden instalar versiones específicas de software.paquetesEsto incluye dependencias, bibliotecas y versiones de Python. Este aislamiento ayuda a evitar conflictos entre versiones de paquetes y garantiza que tus proyectos cuenten con las bibliotecas y herramientas exactas que necesitan. Para obtener más información sobre entornos, para más información puedes consultar [Conceptos sobre entornos](https://www.anaconda.com/docs/getting-started/concepts/what-is-an-environment).

El siguiente comando crea un entorno

```python

conda create --name <ENV_NAME> <PACKAGE>=<VERSION> <PACKAGE> <PACKAGE>
```

```python
### ejemplo de uso  ###

conda create -n mi_entorno python=3.10
```

### Donde:
1. **-n** Asigna el nombre del entorno "mi_entorno".
2. **python=3.10** Instala la versión de Python recomendada para compatibilidad con ROOT.

    <strong>Nota:</strong> Puedes ver todos tus entornos creados con el comando <code>conda env list</code>.

## ACTIVA EL ENTORNO

Dado que los entornos son espacios aislados, solo se puede trabajar con uno a la vez. Seleccionar un entorno para trabajar se denomina activarlo.

```python
conda activate mi_entorno
```

### Perfecto ya tenemos nuestro lugar de trabajo!!!
