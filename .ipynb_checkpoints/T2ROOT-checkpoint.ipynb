## Instalación de ROOT

Puedes descargar el archivo ROOT y escoger la plataforma en la que te encuentras trabajando [Descarga_ROOT](https://root.cern/releases/release-64000-rc1/). Sin ambargo nosotros nos centraremos en una instalación usando nuestro entorno **conda** [ROOT_CONDA](https://root.cern/install/#conda) para evitar errores de librerias o dependencias, aun asi no garantiza que las tengamos a futuro.

---
### CONDA - FORGE

conda-forge es un proyecto comunitario y una organización de GitHub que contiene repositorios de recetas de conda y, por lo tanto, proporciona paquetes de conda para una amplia gama de software.

ROOT se mantiene actualizado en el repositorio **Conda-Forge**.

<div style="background-color: #fff3cd; padding: 10px; border-left: 5px solid #ffecb5;">
    <strong> IMPORTANTE:</strong> Apartir de este punto recuerda mantener tu entorno activado.
</div>


```python
conda config --add channels conda-forge

conda config --set channel_priority strict
```

#### Una vez configurado, podemos instalar **ROOT** desde conda

```python
conda install root -y
```

#### Verificación de instalación


```python
root -l
```

 En este punto ROOT ya debe estar instalado, por lo que incluiremos las librerias necesarias para trabajar

```python
#herramientas de análisis de datos y gráficos

conda install uproot matplotlib numpy pandas -y
```

### Registro del kernel

Para tener un ambiente agradable de trabajo registramos el kernel instalando <code>ipykernel</code> para enlazarlo directamente con <kbd>JupyterNotebook</kbd> y <kbd>VS Code</kbd>

```python
# Instalamos ipykernel
conda install ipykernel -y

# Registramos nuestro entorno para que aparezca en la lista de VS Code

python -m ipykernel install --user --name mi_entorno --display-name "Python (mi_entorno)"
```

### DEPENDENCIAS

A pesar de que hemos instalado todo de manera segurida y completa, agregaremos Dependencias del Sistema, el motor gráfico de ROOT necesita librerías de tu sistema operativo para poder mostrar histogramas y ventanas emergentes [Dependencias](https://root.cern/install/dependencies/#ubuntu-and-other-debian-based-distributions).

```python
sudo apt update

sudo apt install libx11-6 libxpm4 libxft2 libxext6 libglu1-mesa -y

conda install ipywidgets widgetsnbextension -y #  aseguran que  gráficos y tablas funcionen correctamente
```
